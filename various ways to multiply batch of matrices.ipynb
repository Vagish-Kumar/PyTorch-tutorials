```python
"""
Author: Vagish Jha (PhD Scholar, IIT Delhi)
"""

import torch
torch.manual_seed(0)
tensor1 = torch.rand(10,3,5,7)
tensor2 = torch.rand(10,3,7,5)*500

print(f'shape of the tensor1:{tensor1.shape}')
print(f'shape of the tensor2:{tensor2.shape}')
```

```python
#Method-1: Using torch.bmm
output_bmm = torch.empty(10,3,5,5)

for batch in range(tensor1.size(0)):
    print(batch,tensor1.shape,tensor2.shape)
    tensor_a = tensor1[batch,:,:,:]
    tensor_b = tensor2[batch,:,:,:]
    print(tensor_a.shape, tensor_b.shape)
    output = torch.bmm(tensor_a,tensor_b)
    output_bmm[batch] = output
```

```python
print(output_bmm)
```

```python
#Method-2
#using for loop and performing matmul

output_matmul = torch.empty(10,3,5,5)

for batch in range(tensor1.size(0)):
    for channel in range(tensor1.size(1)):
        print(batch,channel)
        tensor_a = tensor1[batch,channel,:,:]
        tensor_b = tensor2[batch,channel,:,:]
        output = torch.matmul(tensor_a,tensor_b)
        print(tensor_a.shape,tensor_b.shape,output.shape)
        output_matmul[batch,channel] = output

#print(tensor_matmul.shape)
print(output_matmul)
```

```python
#Method-3: Using einsum
out_es = torch.einsum('bcij,bcjk->bcik', tensor1,tensor2)
```

```python
#Checking if all the three methods are giving same result
print(torch.mean(output_bmm-output_matmul))
print(torch.mean(output_bmm-out_es))
```

```python
#From the above result, we can see that bmm and einsum is giving exactly same result.
#Also, since there is no loop involved in method 3 (i.e. einsum), therefore, it will be the most efficient. 
```
